```python
import pandas as pd
import numpy as np
```

```python
# Read raw triplet data
triplets_raw = pd.read_csv("things_data48.csv")
# Read triplet names
names = np.loadtxt("things48names.txt", dtype=str, delimiter="\t")
# View data
triplets_raw
```

```python
triplets_raw["item1"] = names[triplets_raw["trip.1"].values - 1]
triplets_raw["item2"] = names[triplets_raw["trip.2"].values - 1]
triplets_raw["item3"] = names[triplets_raw["trip.3"].values - 1]
triplets_raw["chosen_item"] = ""
triplets_raw.loc[triplets_raw["choice"] == 1, "chosen_item"] = triplets_raw.loc[triplets_raw["choice"] == 1, "item1"]
triplets_raw.loc[triplets_raw["choice"] == 2, "chosen_item"] = triplets_raw.loc[triplets_raw["choice"] == 2, "item2"]
triplets_raw.loc[triplets_raw["choice"] == 3, "chosen_item"] = triplets_raw.loc[triplets_raw["choice"] == 3, "item3"]
```

```python
triplets_raw
```

```python
triplets_new = triplets_raw.loc[:, ["worker_no", "exclude", "item1", "item2", "item3", "chosen_item", "isrepeat", "RT"]]
```

```python
triplets_new.to_csv("things48triplets.csv", index=False)
```
